```python
text = sc.textFile('Audio Standardization Sentences.txt',minPartitions=5)
```

```python
words = text.flatMap(lambda x: x.split(' '))
words
```

```python
def count_words(iterator):
    counts = {}
    for w in iterator:
        print(w)
        if w in counts:
            counts[w] += 1
        else:
            counts[w] = 1
    yield counts
```

```python
word_counts = words.mapPartitions(count_words)
```
